```python
def check_payload(message_str):
    msg = email.message_from_string(message_str)
    return bool(msg.get_payload())

def check_payload_length(message_str):
    msg = email.message_from_string(message_str)
    return len(msg.get_payload())

# Apply the function to all messages and count those with payloads
has_payload = df['message'].apply(check_payload)
payload_lengths = df['message'].apply(check_payload_length)
print(f"Emails with payload: {sum(has_payload):,} out of {len(df):,}")
print(f"smallest payload: {min(payload_lengths)}")
```

```python
from presidio_analyzer import AnalyzerEngine
from presidio_anonymizer import AnonymizerEngine
from collections import Counter

def mask_pii_in_email(message_str, entities):
    """
    Takes a raw email message string and a list of PII entity types,
    then analyzes and masks the detected PII (including titles if 'TITLE'
    is in the list of entities).
    """
    # Initialize engines
    analyzer = AnalyzerEngine()
    anonymizer = AnonymizerEngine()
    
    # Get email payload
    msg = email.message_from_string(message_str)
    payload = msg.get_payload()
    
    # Analyze for PII
    analyzer_results = analyzer.analyze(
        text=payload,
        entities=entities,
        language='en'
    )
    
    # Count entities
    entity_counts = Counter(result.entity_type for result in analyzer_results)
    
    # Anonymize the text
    anonymized_text = anonymizer.anonymize(
        text=payload,
        analyzer_results=analyzer_results
    )
    
    return anonymized_text.text, entity_counts

all_base_entities = [
            "PERSON", "PHONE_NUMBER", "EMAIL_ADDRESS", 
            "CREDIT_CARD", "URL", "US_BANK_NUMBER",
            "DATE_TIME", "US_SSN", "US_BANK_NUMBER",
            "US_ITIN", "US_PASSPORT", "LOCATION",
            "IP_ADDRESS", "IBAN_CODE","CRYPTO",
            "MEDICAL_LICENSE"

        ]
# Test it on a specific email
example_masked, counts = mask_pii_in_email(df['message'][495763],all_base_entities)
print("Entity counts:")
for entity, count in counts.items():
    print(f"{entity}: {count}")
print("\nMasked email content:")
print("-" * 80)
print(example_masked)
```

### Lessons

#### Mail Index 496,148 from the Mails
- The US driver’s license is not a reliable metric: [Issue #1063](https://github.com/microsoft/presidio/issues/1063). A better alternative is needed.
  - Found instances of `US_DRIVER_LICENSE: 3` where they should have been categorized as credit cards.
- The code still contains minor errors, such as incorrect categorization of places or names:
  - Example: `VIA SANT'ANTONINO 18 - 50123 FIRENZE` (<LOCATION>)
  - Example: `Paul` and `<PERSON>`
  - Spelling error on `09October` for `<DATE_TIME>`.

#### Mail Index 495,763
- Misclassification of `US_BANK_NUMBER` as a phone number.
- Confusion between `NRP` (Nationality, Religious, or Political group) and `VISA`.
- `US_DRIVER_LICENSE` was mistakenly classified as both a credit card and a phone number.
- Need to mask gendered titles (`Ms.`, `Mr.`) for neutrality.
  - Possible solution: [Deny List Approach](https://microsoft.github.io/presidio/tutorial/01_deny_list/).
  - A better approach might involve replacing titles with gender-neutral alternatives.

#### Speed Test on My PC
- Processing over 500,000 emails will take approximately **350 hours** on my PC.
- A potential improvement: Implement **batch processing**: [Batch Processing Guide](https://microsoft.github.io/presidio/samples/python/batch_processing/).

```python
# Test title with Deny-list
titles_list = [
    "Sir",
    "Ma'am",
    "Madam",
    "Mr.",
    "Mrs.",
    "Ms.",
    "Miss",
    "Dr.",
    "Professor",
]
gender_list = [
    "he",
    "she",
    "him",
    "her",
    "his",
    "hers",
    "himself",
    "herself",
    "men",
    "gentlemen",
    "guys",
    "boys",
    "bros",
    "male",
    "ladies",
    "lady",
    "women",
    "girls",
    "female",
    "chicks",
]
```

```python
from presidio_analyzer import PatternRecognizer

# Create a custom recognizer for TITLE entities using a deny_list.
titles_recognizer = PatternRecognizer(supported_entity="TITLE", deny_list=titles_list)
gender_recognizer = PatternRecognizer(supported_entity="GENDER", deny_list=gender_list)


def mask_pii_in_email_titles_test(message_str, entities):
    """
    Takes a raw email message string and a list of PII entity types,
    then analyzes and masks the detected PII (including titles if 'TITLE'
    is in the list of entities).
    """
    # Initialize engines
    analyzer = AnalyzerEngine()
    anonymizer = AnonymizerEngine()
    
    # Add the custom title recognizer to the analyzer.
    analyzer.registry.add_recognizer(titles_recognizer)
    analyzer.registry.add_recognizer(gender_recognizer)
    
    # Get email payload
    msg = email.message_from_string(message_str)
    payload = msg.get_payload()
    
    # Analyze for PII
    analyzer_results = analyzer.analyze(
        text=payload,
        entities=entities,
        language='en'
    )
    
    # Count entities
    entity_counts = Counter(result.entity_type for result in analyzer_results)
    
    # Anonymize the text
    anonymized_text = anonymizer.anonymize(
        text=payload,
        analyzer_results=analyzer_results
    )
    
    return anonymized_text.text, entity_counts


all_base_entities = [
    "PERSON", "PHONE_NUMBER", "EMAIL_ADDRESS", 
    "CREDIT_CARD", "URL", "US_BANK_NUMBER",
    "DATE_TIME", "US_SSN", "US_ITIN",
    "US_PASSPORT", "LOCATION", "IP_ADDRESS",
    "IBAN_CODE", "CRYPTO", "MEDICAL_LICENSE",
    "TITLE", "GENDER"
]


masked_email, counts = mask_pii_in_email_titles_test(df['message'][495763],all_base_entities)

print("Entity counts:")
for entity, count in counts.items():
    print(f"{entity}: {count}")

print("\nMasked email content:")
print("-" * 80)
print(masked_email)
```
